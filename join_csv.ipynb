```python
import pandas as pd
import sys
```

```python
movies = pd.read_csv('movies.csv')
dates = pd.read_csv('dates.csv')
ratings1 = pd.read_csv('ratings1.csv')
ratings2 = pd.read_csv('ratings2.csv')
ratings = pd.concat([ratings1, ratings2], ignore_index=True)
ratings = ratings.drop_duplicates(ignore_index=True)
ratings_dates = pd.concat([ratings, dates], axis=1)
ratings_dates
```

```python
joined = ratings_dates.join(movies.set_index('movieId'), on='movieId', how='left')
```

```python
ratings_dates.to_csv('ratings.csv', index=False)
```

```python
def join_csv(path1, path2, column_name, join_type='left'):
    file1, file2 = pd.read_csv(path1), pd.read_csv(path2)
    file1.set_index(column_name)
    data = file1.join(file2.set_index(column_name), on=column_name, how=join_type)
    return data
    
join_csv('ratings.csv', 'movies.csv', 'movieId')
    
```
